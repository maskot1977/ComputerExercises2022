```python
def f(x):
    return x**2 + 2 * x + 1
```

```python
X = []
Y = []
for x in range(-3, 3):
    X.append(x)
    Y.append(f(x))

X, Y
```

```python
import matplotlib.pyplot as plt

plt.plot(X, Y)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,4))
plt.title("quadratic function")
plt.plot(X, Y, marker='s', label="y = f(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
import math
def sigma(x):
  return 1 / (1 + math.exp(-x))
```

```python
import math
sigma = lambda x: 1 / (1 + math.exp(-x))
```

```python
X = []
Y = []
for x in range(-100, 101):
  X.append(x/10)
  Y.append(sigma(x/10))
```

```python
import matplotlib.pyplot as plt
plt.plot(X, Y)
plt.grid()
```

```python
import math

g = lambda x: 2**x
h = lambda x: math.log(x)
```

```python
import matplotlib.pyplot as plt

domain = range(1, 10, 2)
X = [x for x in domain]
Y1 = [f(x) for x in X]
Y2 = [g(x) for x in X]
Y3 = [h(x) for x in X]

plt.figure(figsize=(5,4))
plt.title("various functions")
plt.plot(X, Y1, label="y = f(x)", marker="o")
plt.plot(X, Y2, label="y = g(x)", marker="^")
plt.plot(X, Y3, label="y = h(x)", marker="s")
plt.grid()
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
