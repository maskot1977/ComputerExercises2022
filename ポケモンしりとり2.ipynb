```python
pokemon = {}
next_pokemon = {}
small_letters = {
    "ァ": "ア", "ィ":"イ", "ゥ":"ウ", "ェ":"エ", "ォ": "オ", 
    "ャ":"ヤ", "ュ":"ユ", "ョ":"ヨ"
}
for line in open("pokemon_names.txt"):
    name = line.strip()
    top = name[0]
    last = name[-1]
    if last == "ー":
        last = name[-2]
    if last in small_letters.keys():
        last = small_letters[last]
    
    # print(name, top, last)
    pokemon[name] = [top, last]
    if top not in next_pokemon.keys():
        next_pokemon[top] = []
    next_pokemon[top].append(name)
```
