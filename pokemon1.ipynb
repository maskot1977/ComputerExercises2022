```python
def get_last_character(pokemon_name):
    last_character = pokemon_name[-1]
    if last_character == "ー":
        last_character = pokemon_name[-2]
    if last_character == "ャ":
        last_character = "ヤ"
    if last_character == "ュ":
        last_character = "ユ"
    if last_character == "ョ":
        last_character = "ヨ"
    if last_character == "ッ":
        last_character = "ツ"
    if last_character == "ァ":
        last_character = "ア"
    if last_character == "ィ":
        last_character = "イ"
    if last_character == "ゥ":
        last_character = "ウ"
    if last_character == "ェ":
        last_character = "エ"
    if last_character == "ォ":
        last_character = "オ"
    return last_character
```

```python
import copy
class Agent:
    def __init__(self):
        self.path = []
        self.head2names = {}

    def load(self, pokemon_names):
        self.pokemon_names = sorted(pokemon_names)
        for name in pokemon_names:
            head = name[0]
            if head not in self.head2names.keys():
                self.head2names[head] = []
            self.head2names[head].append(name)

    def start_with(self, name):
        self.path.append(name)
    
    def get_index(self, path):
        return [self.pokemon_names.index(name) for name in path]

    def get_next(self, name, ng=None):
        tail = get_last_character(name)
        if tail in self.head2names.keys():
            for c in sorted(list(set(self.head2names[tail]) - set(self.path))):
                if ng is None:
                    return c
                elif self.pokemon_names.index(c) > self.pokemon_names.index(ng):
                    return c
            #print("No more!")
            return None
        else:
            #print(name, tail)
            return False

    def forward(self, num=-1, ng=None):
        next_name = self.get_next(self.path[num], ng=ng)
        if next_name is None:
            #print("No more!!")
            return None
        elif next_name:
            self.path.append(next_name)
            return True
        else:
            return False

longest = []

agent = Agent()
agent.load([line.strip() for line in open("pokemon_names.txt")])
agent.start_with("ピカチュウ")
for iteration in range(int(1e14)):
    #print("iteration", iteration)
    status = agent.forward()
    #print("status", status)
    #if status is None:
    #    print("No more!!!")
    #    agent.path.pop()
    #    agent.path.pop()
    #    agent.forward(ng=agent.path.pop())
    if status:

        if len(longest) < len(agent.path):
            print(iteration, "Longest!")
            longest = copy.deepcopy(agent.path)
            print(len(agent.path), agent.path)
            print()
    else:
        status = agent.forward(ng=agent.path.pop())
        cnt = 0
        while status is None:
            agent.path.pop()
            status = agent.forward(ng=agent.path.pop())
            cnt += 1
            if cnt > 1000:
                break
```
